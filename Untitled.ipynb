```python
# Importing necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Creating the 'Position_Salaries' dataset
data = {
    "Position": ["Intern", "Junior", "Senior", "Lead", "Manager"],
    "Level": [1, 2, 3, 4, 5],
    "Salary": [28000, 43000, 60000, 80700, 100000]
}
position_salaries = pd.DataFrame(data)

# Defining independent variable (Level) and target variable (Salary)
X = position_salaries[["Level"]]  # Independent variable
y = position_salaries["Salary"]  # Target variable

# Splitting data into training and testing sets (7:3 ratio)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Building the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)  # Training the model

# Predicting the test set results
y_pred = model.predict(X_test)

# Printing model parameters and predictions
print("Coefficient (Slope):", model.coef_[0])
print("Intercept:", model.intercept_)
print("Predicted Salaries for Test Set:", y_pred)
print("Mean Squared Error (MSE):", mean_squared_error(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt

# Plotting the training set
plt.scatter(X_train, y_train, color='blue', label='Training Data')

# Plotting the testing set
plt.scatter(X_test, y_test, color='green', label='Testing Data')

# Plotting the regression line
plt.plot(X, model.predict(X), color='red', label='Regression Line')

# Adding labels and legend
plt.title('Position vs. Salary (Linear Regression)')
plt.xlabel('Level (Position)')
plt.ylabel('Salary')
plt.legend()

# Display the plot
plt.show()

```

```python
from sklearn.metrics import r2_score

# Calculating R2 score
r2 = r2_score(y_test, y_pred)
print("R2 Score:", r2)
```
